ANAVO : Analysis of Variance

One-Way Classification

```python
import scipy.stats as stats

stats.f_oneway(dataset['ssc_p'],dataset['hsc_p'],dataset['degree_p'],dataset["etest_p"],dataset["mba_p"])
```

We accept null hypothesis and reject alternate hypothesis

Two-Way Classification

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
edu=['ssc_b','degree_t','salary']
twoway = pd.DataFrame(columns=edu)
```

```python
twoway
```

```python
for columnName in edu:
    twoway[columnName]=dataset[columnName]
    twoway[columnName]=dataset[columnName]
    twoway[columnName]=dataset[columnName]
```

```python
# Perform the two-way ANOVA
model = ols('salary ~ C(ssc_b) + C(degree_t) + C(ssc_b):C(degree_t)', data=twoway).fit()
anova_results = sm.stats.anova_lm(model, typ=2)
```

```python
anova_results
```

```python
p_ssc_b = anova_results.loc['C(ssc_b)', 'PR(>F)']
p_degree_t = anova_results.loc['C(degree_t)', 'PR(>F)']
p_interaction = anova_results.loc['C(ssc_b):C(degree_t)', 'PR(>F)']

print(f"P-value for 10th board: {p_ssc_b:.6f}")
print(f"P-value for degree: {p_degree_t:.6f}")
print(f"P-value for interaction: {p_interaction:.6f}")
```

```python
#from the above results there is no significant so Accept Null Hypothesis and reject Alternate Hypothesis
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Create the data
df = pd.DataFrame({
    'water': np.repeat(['daily', 'weekly'], 15),
    'sun': np.tile(np.repeat(['low', 'med', 'high'], 5), 2),
    'height': [6, 6, 6, 5, 6, 5, 5, 6, 4, 5, 6, 6, 7, 8, 7, 3, 4, 4, 4, 5, 4, 4, 4, 4, 4, 5, 6, 6, 7, 8]
})

# Perform the two-way ANOVA
model = ols('height ~ C(water) + C(sun) + C(water):C(sun)', data=df).fit()
# Get the ANOVA table
anova_results = sm.stats.anova_lm(model, typ=2)

# Extract p-values
p_water = anova_results.loc['C(water)', 'PR(>F)']
p_sun = anova_results.loc['C(sun)', 'PR(>F)']
p_interaction = anova_results.loc['C(water):C(sun)', 'PR(>F)']

print(f"P-value for water: {p_water:.6f}")
print(f"P-value for sun: {p_sun:.6f}")
print(f"P-value for interaction: {p_interaction:.6f}")
```

```python
#p-value of water and sun is significant so we can reject null hypothesis and accept alternate hypothesis
```

```python
#p-value of interaction is not significant so we can accept null hypothesis and reject alternate hypothesis
```

Here, weâ€™re creating a linear regression model using the ols (ordinary least squares) function from the statsmodels library.

The formula 'height ~ C(water) + C(sun) + C(water):C(sun)' specifies the dependent variable (height) and the independent variables (water and sun), including their interaction.

The C() function indicates that these variables are categorical.

The data=df argument specifies the DataFrame (df) containing the data.

.fit() fits the model to the data.

After fitting the model, we perform an ANOVA to analyze the variance.

The typ=2 argument specifies the type of sum of squares (Type 2).

The results include F-statistics, p-values, and other relevant information.
